```python
import numpy as np
import pandas as pd
import yfinance as yf
from datetime import timedelta

import mplfinance as mpf
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
ohlc_1d = yf.download('^BVSP', start='2015-01-01', end='2023-01-01')
ohlc_1d['MM_20p'] = ohlc_1d.Close.rolling(20).mean()
ohlc_1d['Dist_close_mm'] = abs((ohlc_1d['Close'] - ohlc_1d['MM_20p'])/ohlc_1d['MM_20p'])
ohlc_1d
```

```python
#dados_fig = ohlc_1d
dados_fig = ohlc_1d.loc['2016-08-01':'2016-09-01']

fig = make_subplots(rows=2, cols=1)

fig.add_trace(go.Candlestick(name='Ativo',x=dados_fig.index, close=dados_fig['Close'], open=dados_fig['Open'],
                             high=dados_fig['High'], low=dados_fig['Low']), row=1,col=1)

fig.add_trace(go.Scatter(name='Media movel', x=dados_fig.index, y=dados_fig['MM_20p'],
              marker_color='blue'), row=1,col=1)

fig.add_trace(go.Scatter(name='Distancia Close vs MM', x=dados_fig.index, y=dados_fig['Dist_close_mm'],
              marker_color='purple'), row=2, col=1)

fig.update_xaxes(rangebreaks=[dict(bounds=["sat", "mon"])])

fig.update_layout(title='<b>Entendendo o Power Breakout', xaxis2_title="<b>Data", yaxis1_title="<b>Valor nominal do ativo",
                  yaxis2_title="<b>Distancia Close vs MM",
                  xaxis1_rangeslider_visible=False,
                  legend=dict(orientation="v",yanchor="top",y=1, xanchor="right",x=1.30),
                  template = 'simple_white', paper_bgcolor="#f7f8fa", margin=dict(l=20, r=20, t=50, b=20),
                  font=dict(family="Arial",size=15,color="black"),
                  height=600, width=1000)

fig.show()
```

```python
ohlc_1d['Variacao_close'] = abs((ohlc_1d['Close'] - ohlc_1d['Close'].shift(1))/ohlc_1d['Close'].shift(1))
ohlc_1d['Variacao_MM'] = abs((ohlc_1d['MM_20p'] - ohlc_1d['MM_20p'].shift(1))/ohlc_1d['MM_20p'].shift(1))
```

```python
dados_fig = ohlc_1d.loc['2022-07-10':'2022-09-10']

fig = make_subplots(rows=2, cols=1)

fig.add_trace(go.Candlestick(name='Ativo',x=dados_fig.index, close=dados_fig['Close'], open=dados_fig['Open'],
                             high=dados_fig['High'], low=dados_fig['Low']), row=1,col=1)
fig.add_trace(go.Scatter(name='Media movel', x=dados_fig.index, y=dados_fig['MM_20p'],
              marker_color='blue'), row=1,col=1)


fig.add_trace(go.Scatter(name='Distancia Close vs MM', x=dados_fig.index, y=dados_fig['Dist_close_mm'],
              marker_color='purple'), row=2, col=1)
fig.add_trace(go.Scatter(name='Variacao Close', x=dados_fig.index, y=dados_fig['Variacao_close'],
              marker_color='red'), row=2, col=1)
fig.add_trace(go.Scatter(name='Variacao MM', x=dados_fig.index, y=dados_fig['Variacao_MM'],
              marker_color='blue'), row=2, col=1)


fig.update_xaxes(rangebreaks=[dict(bounds=["sat", "mon"])])

fig.update_layout(title='<b>Entendendo o Power Breakout', xaxis2_title="Data", yaxis1_title="Valor nominal do ativo",
                  yaxis2_title="Distancia Close vs MM",
                  xaxis1_rangeslider_visible=False,
                  legend=dict(orientation="v",yanchor="top",y=1, xanchor="right",x=1.30),
                  template = 'simple_white', paper_bgcolor="#f7f8fa", margin=dict(l=20, r=20, t=50, b=20),
                  font=dict(family="Arial",size=15,color="black"),
                  height=600, width=1000)

fig.show()
```

```python
ohlc_1d['Power_breakout'] = (ohlc_1d['Variacao_MM'] > ohlc_1d['Variacao_close']) & \
                            (ohlc_1d['Variacao_MM'].shift(1) > ohlc_1d['Variacao_close'].shift(1)) & \
                            (ohlc_1d['Variacao_MM'].shift(2) > ohlc_1d['Variacao_close'].shift(2))# & \
                            #(ohlc_1d['Dist_close_mm'] < 0.02)
```

```python
df_teste_indicador = ohlc_1d[ohlc_1d['Power_breakout']==True]
df_teste_indicador
```

# Usando médias das médias

```python
ohlc_1d['MM_var_close'] = ohlc_1d['Variacao_close'].rolling(20).mean()
ohlc_1d['MM_var_MM'] = ohlc_1d['Variacao_MM'].rolling(20).mean()
ohlc_1d['Diff_mm'] = ohlc_1d['MM_var_close'] - ohlc_1d['MM_var_MM']
```

```python
#dados_fig = ohlc_1d.loc['2017-09-01':'2017-10-01']
dados_fig = ohlc_1d

fig = make_subplots(rows=2, cols=1)

fig.add_trace(go.Candlestick(name='Ativo',x=dados_fig.index, close=dados_fig['Close'], open=dados_fig['Open'],
                             high=dados_fig['High'], low=dados_fig['Low']), row=1,col=1)
fig.add_trace(go.Scatter(name='Media movel', x=dados_fig.index, y=dados_fig['MM_20p'],
              marker_color='blue'), row=1,col=1)

fig.add_trace(go.Scatter(name='Distancia Close vs MM', x=dados_fig.index, y=dados_fig['MM_var_close'],
              marker_color='purple'), row=2, col=1)
fig.add_trace(go.Scatter(name='Variacao Close', x=dados_fig.index, y=dados_fig['MM_var_MM'],
              marker_color='red'), row=2, col=1)
fig.add_trace(go.Scatter(name='Variacao Close', x=dados_fig.index, y=dados_fig['Diff_mm'],
              marker_color='green'), row=2, col=1)

fig.update_xaxes(rangebreaks=[dict(bounds=["sat", "mon"])])

fig.update_layout(title='<b>Entendendo o Power Breakout', xaxis2_title="Data", yaxis1_title="Valor nominal do ativo",
                  yaxis2_title="Distancia Close vs MM",
                  xaxis1_rangeslider_visible=False,
                  legend=dict(orientation="v",yanchor="top",y=1, xanchor="right",x=1.30),
                  template = 'simple_white', paper_bgcolor="#f7f8fa", margin=dict(l=20, r=20, t=50, b=20),
                  font=dict(family="Arial",size=15,color="black"),
                  height=600, width=1000)

fig.show()
```
