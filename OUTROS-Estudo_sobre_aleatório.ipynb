```python
dados = {}
qtde_simulacoes = 5
qtde_elementos = 1000000

for simulação in range(0,qtde_simulacoes): # rodar a qtde de simulações definida
  dados['simulacao_' + str(simulação)] = [] # cria as keys do dicionário
  for elemento in range(0,qtde_elementos): # rodar a qtde de elemtentos definidos
    dados['simulacao_' + str(simulação)].append(random.choice([-1, 1])) # apenda os valores na key

df = pd.DataFrame(dados)

for i in range(0,qtde_simulacoes):

  df['soma_simulacao_' + str(i)] = df['simulacao_' + str(i)].cumsum()

df.iloc[:, qtde_simulacoes:].plot(figsize=(7,5));
```

# Estudo Antigo

#### Estudo com aleatório binário (-1 e 1)

```python
# Definir qtde de elementos aleatórios que serão criados
qtde_elementos = 1000

# Apendando os numeros gerados em uma lista e transformando-a em DataFrame
lista_numeros = []

for i in range(0,(qtde_elementos)):
  lista_numeros.append(random.choice([-1,1]))
  
  
df_lista_numeros = pd.DataFrame(lista_numeros, columns=['numero'])




# Por fim, vamos criar uma coluna com o acumulado dos números gerados e plotar:
df_lista_numeros['soma'] = df_lista_numeros['numero'].cumsum()
df_lista_numeros['media_20'] = df_lista_numeros['soma'].rolling(20).mean()
df_lista_numeros['media_200'] = df_lista_numeros['soma'].rolling(200).mean()
df_lista_numeros.plot(figsize=(7,5), y=['soma', 'media_20','media_200']);
```

#### Estudo com numeros aleatórios de -10 a 10

```python
qtde_ele = 1000

list = []
for i in range(0,(qtde_ele)):
  list.append(random.randint(-10,10))
df = pd.DataFrame(list, columns=['-10_a_10'])

df['soma'] = df['-10_a_10'].cumsum()
df['media_20'] = df['soma'].rolling(20).mean()
df['media_200'] = df['soma'].rolling(200).mean()
df.plot(figsize=(15,7), y=['soma', 'media_20','media_200']);
```
