```python
import yfinance as yf
import mplfinance as mpf
```

```python
ibov = yf.download('^BVSP', start='2008-01-01')
```

```python
mpf.plot(ibov, type='candle', style="charles")
```

```python
ibov.Low.min()
```

```python
ibov.High.max()
```

```python
ibov.Close.iloc[-1]
```

```python
round(ibov.High.max()/ibov.Low.min(), 3) - 1
```

```python
round(ibov.Close.iloc[-1]/ibov.Low.min(), 3) - 1
```

```python
dolar = yf.download('BRL=X', start='2008-01-01')
```

```python
ibov_dolarizado = ibov/dolar
```

```python
ibov_dolarizado = ibov_dolarizado.dropna()
```

```python
mpf.plot(ibov_dolarizado, type='candle', style="charles")
```

```python
ibov_dolarizado.Low.min()
```

```python
ibov_dolarizado.High.max()
```
